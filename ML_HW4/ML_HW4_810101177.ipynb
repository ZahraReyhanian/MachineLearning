```python
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
from sklearn import svm
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import seaborn as sns; sns.set()
```

```python
df = px.data.iris()
df.head()
```

```python
np.unique(df['species'])
```

```python
X = df[['petal_length', 'petal_width']]
y = np.array(df['species'].map({'setosa': 2, 'versicolor': 1, 'virginica': 0}))
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.6, random_state=0)
```

```python
# C is the SVM regularization parameter
C = 10.0 
  
# Create an Instance of SVM and Fit out the data.
# Data is not scaled so as to be able to plot the support vectors
svc = svm.SVC(kernel ='linear', C = C, decision_function_shape='ovr').fit(X_train, y_train)
```

```python
# create a mesh to plot
x_min, x_max = X.iloc[:, 0].min() - 1, X.iloc[:, 0].max() + 1
y_min, y_max = X.iloc[:, 1].min() - 1, X.iloc[:, 1].max() + 1
h = 0.03
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
         np.arange(y_min, y_max, h))
```

```python
# Plot the data for Proper Visual Representation
plt.subplot(1, 1, 1)
  
# Predict the result by giving Data to the model
Z = svc.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, cmap = plt.cm.Paired, alpha = 0.8)
  
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c = y, cmap = plt.cm.Paired)
plt.xlabel('Petal length')
plt.ylabel('Petal width')
plt.xlim(xx.min(), xx.max())
plt.title('SVM with linear kernel, one vs rest')
  
# Output the Plot
plt.show()
```

```python
y_pred = svc.predict(X_train)
accuracy = accuracy_score(y_train, y_pred)
print(f'Accuracy in train data: {accuracy}')

y_pred = svc.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy in test data: {accuracy}')

confusion_mtx = confusion_matrix(y_test, y_pred)
sns.heatmap(confusion_mtx, square=True, annot=True, fmt='d', cbar=True);
```

```python
import random
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
```

```python
X = np.linspace(0, 2, 100)
er = np.random.random_sample(size = 100)/2 - 0.5

y = np.sin(X**2) + er
plt.scatter(X, y, s=15)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
X = X.reshape(100,1)
y = y.reshape(100,1)

sc_X = StandardScaler()
sc_y = StandardScaler()
X = sc_X.fit_transform(X)
y = sc_y.fit_transform(y)
```

```python
reg_rbf = SVR(kernel = 'rbf')

parameters={'C':[1e-15, 1e-10, 1e-8, 1e-3, 1e-2, 1, 5, 10, 20, 30, 35, 40, 45, 50, 55, 100],
               'gamma': [1e-15, 1e-10, 1e-8, 1e-3, 1e-2, 1, 5, 10, 100]}
reg_ridge = GridSearchCV(reg_rbf, parameters, scoring='neg_mean_squared_error', cv=5)
reg_ridge.fit(X, y)

print(reg_ridge.best_params_)
print(reg_ridge.best_score_)
```

```python
X_grid = np.arange(min(X), max(X), 0.01) #this step required because data is feature scaled.
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, s = 15)
plt.plot(X_grid, reg_ridge.predict(X_grid), color = 'red')
plt.title('SVR  with RBF Kernel')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
reg_poly = SVR(kernel = 'poly', degree=3)

parameters={'C':[1e-15, 1e-10, 1e-8, 1e-3, 1e-2, 1, 5, 10, 20, 30, 40, 50, 100],
               'gamma': [1e-15, 1e-10, 1e-8, 1e-3, 1e-2, 1]}
reg_ridge = GridSearchCV(reg_poly, parameters, scoring='neg_mean_squared_error', cv=5)
reg_ridge.fit(X, y)

print(reg_ridge.best_params_)
print(reg_ridge.best_score_)
```
